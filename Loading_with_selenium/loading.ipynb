```python
from selenium.webdriver import Chrome
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
from fake_useragent import UserAgent
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
import time
import pandas as pd
```

```python
options = Options()
#options.headless = True
ua = UserAgent()
userAgent = ua.random
options.add_argument(f'user-agent={userAgent}')


driver = webdriver.Chrome(r'C:\Users\HP\Downloads\Premier_League_Stats\Loading_with_selenium\chromedriver.exe',options=options)

def get_players():
    driver.get("https://www.premierleague.com/players")
    SCROLL_PAUSE_TIME = 20

    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    content = []      
    contents =driver.find_elements_by_css_selector("div table tbody tr")
    for con in contents:
        cont = con.get_attribute('innerText')
        cont = cont.replace('\n\t',',')
        cont = cont.replace('\t',',')
        content.append(cont)
        
    af = pd.DataFrame(content,columns =['content'])
    af = af['content'].str.split(",",n = 3, expand = True)
    af.columns = ['player','position','country']
    af.to_csv('./csv_dir/player.csv',index = False)
    

    

def get_clubs():
    driver.get("https://www.premierleague.com/clubs")
    content = []      
    contents =driver.find_elements_by_class_name('nameContainer')
    for con in contents:
        cont = con.get_attribute('innerText')
        cont = cont.replace('\n',',')
        content.append(cont)
    af = pd.DataFrame(content,columns =['content'])
    af = af['content'].str.split(",",n = 2, expand = True)
    af.columns = ["club_name","stadium"]
    af.to_csv('./csv_dir/club.csv',index = False)

def get_managers():
    driver.get("https://www.premierleague.com/managers")
    time.sleep(5)
    content = []      
    contents =driver.find_elements_by_class_name('activeManager')
    for con in contents:
        cont = con.get_attribute('innerText')
        cont = cont.replace('\t',',')
        cont = cont.replace('\n','')
        content.append(cont)
    af = pd.DataFrame(content,columns =['content'])
    af = af['content'].str.split(",",n = 3, expand = True)
    af.columns = ["manager_name","club",'nationality'] 
    af.to_csv('./csv_dir/manager.csv',index = False)
    
get_players()
get_clubs()
get_managers()

driver.quit()

```
